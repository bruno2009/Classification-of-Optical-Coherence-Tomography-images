```python
import numpy as np
import pandas as pd

import pandas as pd
import numpy as np
from plotnine import *
from mizani.formatters import percent_format

pd.set_option('display.width', 180)

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel('Result_knn_CV_Final.xlsx', sheet_name=0)
melted_df = pd.melt(df, id_vars=['folds'], value_vars=['split0_test_score', 'split1_test_score', 'split2_test_score', 'split3_test_score'], var_name='Split', value_name='value')
```

```python
points = pd.DataFrame( {'x': [np.nan], 'y': [np.nan], 'Mean': ['Mean']} )

(
    ggplot(melted_df, aes(x='folds', y='value' ))
    + geom_point()
    + geom_point(points, aes(x='x', y = 'y', color = 'Mean' ) )
    + stat_summary(fun_y = np.mean,  ymax = 0, ymin = 0, color = 'red')
    + scale_color_manual(name = ' ', values= ['red'])
    + scale_y_continuous(name = " ", breaks = np.arange(0.7, 1.0, 0.05), limits=(0.7, 1.0))
    + scale_x_continuous(name = "Candidates from GridSearchCV", breaks = np.arange(0, 9, 1), limits=(0, 8))
    + ggtitle("Cohen kappa score")
    + theme_bw()
    +   theme(axis_line = element_line(size=1, colour = "black"),
        plot_title = element_text(size = 15, family="Comic Sans MS", hjust = 0),  # hjust = 0.5 (center) and hjust = 1 (right)
        text = element_text(size = 16, family="Comic Sans MS"),
        axis_text_x = element_text(colour="black", size = 12.5, face = "italic"),
        axis_text_y = element_text(colour="black", size = 12.5),
        axis_title_y = element_text(size=12.5), 
        axis_title_x = element_text(size=12.5),
        figure_size=(8, 4), 
        legend_key=element_rect(fill="white"),
        legend_text = element_text(ha= "left", size=12), 
        legend_background = element_rect(fill = None), 
        # legend_position="bottom",
        legend_title_align = 'left',
        legend_direction = "horizontal",
        legend_box_spacing = 0.01
        # legend_position= (0.2, 0.8)
        )
)
```
